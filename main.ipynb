```python
from sklearn.datasets import load_digits
import module as ml
import visualization as vis
# сделать модули, подгрузка модулей, расширение .py
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
# from sklearn.metrics import accuracy_score
```

```python
# Загрузка датасета
data = load_digits()
X, y = ml.load_data(data)
```

```python
# Разделение данных на обучающую и тестовую выборки, стратификация по y
X_train, X_test, y_train, y_test = ml.separation(X, y)
```

```python

# Нормализация данных
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Создание и обучение модели логистической регрессии
y_pred = ml.education(X_train, y_train, X_test)
```

```python
# оценка точности модели
ml.model_accuracy(y_test, y_pred)
```

```python
vis.plot_confusion_matrix(y_test, y_pred, class_labels=[str(i) for i in range(10)])
```
